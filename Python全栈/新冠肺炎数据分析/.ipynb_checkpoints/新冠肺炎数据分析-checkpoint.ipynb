```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams.update({'font.size':14})
```

```python
# parse_dates参数：将csv中的时间字符串转换成日期格式，会减缓读取速度
data = pd.read_csv('./novel-corona-virus-2019-dataset/covid_19_data.csv', 
                   parse_dates = ['ObservationDate', 'Last Update'])
data.head()
```

```python
data.info()
```

```python
# 删除不必要的列
data = data.drop(['SNo','Last Update'], axis = 1)
data = data.rename(columns={'Country/Region': 'Country', 'ObservationDate': 'Date'})
```

```python
data.head()
```

```python
# 空值的地区
data.isnull().sum()
```

```python
# 按-国家-省-日期分组 筛选出只有一例观察的地区
checkdup = data.groupby(['Country', 'Province/State','Date']).count().iloc[:,0]
```

```python
checkdup[checkdup > 1]
```

```python
checkdup
```

```python
daily = data.sort_values(['Date', 'Country', 'Province/State'])
daily.head(20)
```

```python
# 区分湖北地区、中国地区和其他地区
def get_place(row):
    if row['Province/State'] == 'Hubei':
        return 'Hubei PRC'
    elif row['Country'] == 'Mainland China':
        return 'Others PRC'
    else:
        return 'World'

daily['segment'] = daily.apply(lambda row: get_place(row), axis=1)
```

```python
daily.head(20)
```

## 最新疫情数据

```python
latest = daily[daily.Date == daily.Date.max()]
latest.shape 
```

```python
print('确诊病例总数：%.d' % np.sum(latest['Confirmed']))
print('死亡病例总数：%.d' % np.sum(latest['Deaths']))
print('治愈病例总数：%.d' % np.sum(latest['Recovered']))
```

## 统计地区确诊、死亡、治愈、死亡率、治愈率

```python
segment1 = latest.groupby('segment').sum()
segment1['Death Rate'] = segment1['Deaths'] / segment1['Confirmed'] * 100
segment1['Recovered Rate'] = segment1['Recovered'] / segment1['Confirmed'] * 100
```

```python
segment1
```

### 发现：
1. 湖北省确诊病例比中国其他省份的总和还要多。

2. 绝大多数死亡病例来自湖北省。

3. 湖北省的死亡率较高，为4.3%，其他中华人民共和国和世界各国的死亡率较低，分别为0.85%和1.8%。

4. 中国的治愈率似乎要高得多，湖北为57%，其他地方为88%。

## 定义通用绘图函数

```python
def barplot_show(x, y, data=None, title:str=None):
    plt.figure(figsize=(9,6))
    sns.barplot(x, y, data=data)
    plt.title(title)
    plt.yticks(fontsize=12)
    plt.grid(axis='x')
    plt.show()
```

## 中国湖北以外地区

```python
# Confirmed Cases China ex-Hubei
China_other_province_confirmed_df = latest.loc[latest.segment=='Others PRC',['Province/State','Confirmed']].sort_values('Confirmed', ascending=False)
plt.figure(figsize=(9,7))
sns.barplot('Confirmed', 'Province/State', data = China_other_province_confirmed_df)
plt.title('Top confirmed cases China ex-Hubei')
plt.yticks(fontsize=12)
plt.grid(axis='x')
plt.show()
```

```python
China_others_province_latest_deaths_df = latest.loc[latest.segment=='Others PRC', ['Province/State', 'Deaths']].sort_values('Deaths', ascending=False)
China_others_province_latest_deaths_df = China_others_province_latest_deaths_df[China_others_province_latest_deaths_df.Deaths > 0]
# 中国所有地区死亡数
# China_all_province_latest_deaths_df = latest.loc[latest.segment.isin(['hubei PRC','Others PRC']), ['Province/State', 'Deaths']].sort_values('Deaths', ascending=False)
```

```python
# 中国湖北以外地区死亡病例数
plt.figure(figsize=(9,7))
sns.barplot('Deaths', 'Province/State', data = China_others_province_latest_deaths_df)
plt.title('Top Death cases China ex-Hubei')
plt.yticks(fontsize=12)
plt.grid(axis='x')
plt.show()
```

```python
# 除中国外确诊病例数
world_latest_confirmed_df = latest.loc[latest.segment=='World', ['Country', 'Confirmed']]
world_latest_confirmed_df = world_latest_confirmed_df.groupby('Country').sum().sort_values('Confirmed',ascending=False)
top10_world_latest_confirmed_df = world_latest_confirmed_df.head(10)
top10_world_latest_confirmed_df
```

```python
# plt.figure(figsize=(9,6))
# sns.barplot(top10_world_latest_deaths_df.values, top10_world_latest_deaths_df.index)
# plt.title('Top 10 Confirmed cases World ex-China')
# plt.yticks(fontsize=12)
# plt.grid(axis='x')
# plt.show()
barplot_show(x=top10_world_latest_confirmed_df.Confirmed, y=top10_world_latest_confirmed_df.index,
            title='Top 10 Confirmed cases World ex-China')
```

```python
# 中国以外国家死亡病例数
world_latest_deaths_df = latest.loc[latest.segment=='World', ['Country', 'Deaths']].groupby('Country').sum()
top15_world_latest_deaths_df = world_latest_deaths_df.sort_values('Deaths', ascending=False).head(15)
top15_world_latest_deaths_df
```

```python
barplot_show(top15_world_latest_deaths_df['Deaths'], 
             top15_world_latest_deaths_df.index,
            title='Top 15 Deaths cases World ex-China')
```

```python
_ = latest.groupby('Country')['Confirmed', 'Deaths'].sum().reset_index()
_['Death rate'] = _['Deaths'] / _['Confirmed'] * 100
_ = _.sort_values('Death rate', ascending=False)
_ = _[_['Deaths'] > 0]
barplot_show('Death rate', 'Country', data=_, title='Top Death Rate Whole World')
```

## 发现：
- 菲律宾是目前为止死亡率最高的国家(但只有三分之一)

- 美国的死亡率已经排名第二

- 中国大陆，此次爆发的中心，排名第六

- 尽管案件数量激增，但韩国的死亡率很低

```python
confirm = pd.pivot_table(daily.dropna(subset=['Confirmed']), index='Date', 
                         columns='segment', values='Confirmed', aggfunc=np.sum).fillna(method='ffill')
plt.figure(figsize=(11,6))
plt.plot(confirm, marker='o')
plt.title('Confirmed Cases')
plt.legend(confirm.columns, loc=2, fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## 发现：

- 中国最近的新病例要少得多

- 中国以外的病例已经超过了中国(湖北除外)

```python
death = pd.pivot_table(daily.dropna(subset=['Deaths']), index='Date',
                      columns='segment', values='Deaths', aggfunc=np.sum).fillna(method='ffill')
plt.figure(figsize=(11,6))
plt.plot(death, marker='o')
plt.title('Death Cases')
plt.legend(death.columns, loc=2, fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## 发现：
- 世界死亡病例已超过中国(除湖北外)。

```python
good = pd.pivot_table(daily.dropna(subset=['Recovered']), 
                         index='Date', columns='segment', values='Recovered', aggfunc=np.sum).fillna(method = 'ffill')
plt.figure(figsize=(11,6))
plt.plot(good, marker='o')
plt.title('Recovered Cases')
plt.legend(good.columns, loc=2, fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# 存活病例——确诊-死亡-治愈
daily['Active'] = daily['Confirmed'] - daily['Deaths'] - daily['Recovered']
active = pd.pivot_table(daily.dropna(subset=['Active']), index='Date', 
                        columns='segment', values='Active', aggfunc=np.sum).fillna(method='ffill')
plt.figure(figsize=(11,6))
plt.plot(active, marker='o')
plt.title('Active Cases')
plt.legend(active.columns, loc=2, fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## 发现：

- 自2月中旬以来，湖北的存活病例有所下降

- 在中国其他地区，这个数字接近于0

```python
top10 = top10_world_latest_confirmed_df.index.tolist()
top10_cases = daily[daily['Country'].isin(top10)]
world_confirmed_daily = pd.pivot_table(top10_cases.dropna(subset=['Confirmed']), index='Date',
                                columns='Country', values='Confirmed', aggfunc=np.sum).fillna(method='ffill')
world_confirmed_daily
```

```python
# 2月1号之前，大部分国家尚未有确诊病例，所以我们从2月1号开始作图
plt.figure(figsize=(12,7))
plt.plot(world_confirmed_daily[world_confirmed_daily.index > '2020-02-01'], marker='o')
plt.title('Confirmed Cases Daily - Top 10 Contries Outside China')
plt.legend(world_confirmed_daily.columns, loc=2, fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df = confirm.join(death, lsuffix='_confirm', rsuffix='_death')
df = df.join(good.add_suffix('_recover'))
df.head()
```

```python
df['Hubei PRC_death_rate'] = df['Hubei PRC_death']/df['Hubei PRC_confirm']
df['Others PRC_death_rate'] = df['Others PRC_death']/df['Others PRC_confirm']
df['World_death_rate'] = df['World_death']/df['World_confirm']
df['Hubei PRC_recover_rate'] = df['Hubei PRC_recover']/df['Hubei PRC_confirm']
df['Others PRC_recover_rate'] = df['Others PRC_recover']/df['Others PRC_confirm']
df['World_recover_rate'] = df['World_recover']/df['World_confirm']
```

```python
death_rate = df[['Hubei PRC_death_rate','Others PRC_death_rate','World_death_rate']]*100
plt.figure(figsize=(11,6))
plt.plot(death_rate, marker='o')
plt.title('Death Rate %')
plt.legend(death.columns)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
recover_rate = df[['Hubei PRC_recover_rate','Others PRC_recover_rate','World_recover_rate']]*100
plt.figure(figsize=(11,6))
plt.plot(recover_rate, marker='o')
plt.title('Recovery Rate %')
plt.legend(good.columns, loc=2, fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
