```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd
import random
import math
import time
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error
import datetime
%matplotlib inline
```

```python
confirmed_df = pd.read_csv("./novel-corona-virus-2019-dataset/time_series_covid_19_confirmed.csv")
deaths_df = pd.read_csv("./novel-corona-virus-2019-dataset/time_series_covid_19_deaths.csv")
recoveries_df = pd.read_csv("./novel-corona-virus-2019-dataset/time_series_covid_19_recovered.csv")
```

```python
confirmed_df.head()
```

```python
cols = confirmed_df.keys()
```

```python
cols
```

```python
# 确诊数，从2020/1/22号至今
confirmed = confirmed_df.loc[:, cols[4]:cols[-1]]
deaths = deaths_df.loc[:, cols[4]:cols[-1]]
recoveries = recoveries_df.loc[:, cols[4]:cols[-1]]
```

```python
# 求每一天确诊数之和、死亡数之和以及死亡率、治愈数之和
dates = confirmed.keys()
world_cases = []
total_deaths = []
mortality_rate = []    # 死亡率
total_recovered = []

for i in dates:
    confirmed_sum = confirmed[i].sum()
    death_sum = deaths[i].sum()
    recovered_sum = recoveries[i].sum()
    world_cases.append(confirmed_sum)
    total_deaths.append(death_sum)
    mortality_rate.append(death_sum/confirmed_sum)
    total_recovered.append(recovered_sum)

# 列表转numpy矩阵
days_since_1_22 = np.array([i for i in range(len(dates))]).reshape(-1, 1)  # 30天
world_cases = np.array(world_cases).reshape(-1, 1)            # 确诊数
total_deaths = np.array(total_deaths).reshape(-1, 1)          # 死亡数
total_recovered = np.array(total_recovered).reshape(-1, 1)    # 治愈数
```

## 特征预测

```python
# 预测3天
days_in_future = 3
future_forcast = np.array([i for i in range(len(dates)+days_in_future)]).reshape(-1, 1)
```

### 整数转时间序列便于可视化

```python
start = '1/22/2020'
start_date = datetime.datetime.strptime(start, '%m/%d/%Y')
future_forcast_dates = []
for i in range(len(future_forcast)):
    future_forcast_dates.append((start_date + datetime.timedelta(days=i)).strftime('%m/%d/%Y'))
adjusted_dates = future_forcast_dates[:-3]
```

### 前80%天的数据作为训练集，后20%天的作为测试集

```python
# X为索引，y为确诊数
X_train_confirmed, X_test_confirmed, y_train_confirmed, y_test_confirmed = train_test_split(days_since_1_22, world_cases, test_size=0.2, shuffle=False)
```

## 确认病例数预测模型。在这个例子中，我使用了支持向量机、随机森林和线性回归。

### SVM预测

```python
kernel = ['linear', 'rbf']
c = [0.01, 0.1, 1, 10]
gamma = [0.01, 0.1, 1]
epsilon = [0.01, 0.1, 1]
shrinking = [True, False]
svm_grid = {'kernel': kernel, 'C': c, 'gamma': gamma, 'epsilon': epsilon, 
            'shrinking': shrinking}
svm = SVR()
svm_search = RandomizedSearchCV(svm, svm_grid, scoring='neg_mean_squared_error', cv=3, 
                                return_train_score=True, n_jobs=-1, n_iter=30, verbose=1)
svm_search.fit(X_train_confirmed, y_train_confirmed)
```

```python
svm_search.best_params_
```

```python
svm_confirmed = svm_search.best_estimator_
svm_confirmed
```

```python
# 使用SVM预测
svm_test_pred = svm_confirmed.predict(X_test_confirmed)   # 预测测试集
svm_pred = svm_confirmed.predict(future_forcast)          # 预测整个集合
# 预测走势，蓝色线
plt.plot(X_test_confirmed, svm_test_pred, linewidth=3, color='blue')
plt.plot(svm_pred, linestyle='--', color='green')
# 实际走势，红色线
plt.plot(world_cases, color='red') 
plt.show()
```

```python
# 平均绝对值误差
print("MAE:", mean_absolute_error(svm_test_pred, y_test_confirmed))
# 均方差误差，扩大了预测错误的惩罚
print("MSE:", mean_squared_error(svm_test_pred, y_test_confirmed))
```

#### 发现：
- 实际走势为红线，预测走势为虚线和蓝实线
- SVM和实际走势差别较大

### 线性回归预测

```python
linear_model = LinearRegression(normalize=True)
linear_model.fit(X_train_confirmed, y_train_confirmed)
linear_test_pred = linear_model.predict(X_test_confirmed)
linear_pred = linear_model.predict(future_forcast)
plt.plot(X_test_confirmed, linear_test_pred, linewidth=4, color='blue')
plt.plot(linear_pred, linestyle='--', color='green')
plt.plot(world_cases, color='red')
plt.show()
```

```python
print("MAE:", mean_absolute_error(linear_test_pred, y_test_confirmed))
print("MSE:", mean_squared_error(linear_test_pred, y_test_confirmed))
```

```python
print(linear_model.coef_)        # 回归系数，如一元模型里 y=Ax+B 里的 A
print(linear_model.intercept_)   # 截距，如一元模型里 y=Ax+B 里的 B
```

### 获取有关已确诊冠状病毒病例的国家/地区的信息

#### 国家最新

```python
unique_countries =  list(confirmed_df['Country/Region'].unique())
```

```python
country_confirmed_cases = {}
for i in unique_countries:
    cases = latest_confirmed[confirmed_df['Country/Region']==i].sum()
    if cases > 0:
        country_confirmed_cases[i] = cases
    else:
        unique_countries.remove(i)
```

```python
country_confirmed_cases = sorted(country_confirmed_cases.items(), key=lambda x:x[1], reverse=True)
country_confirmed_cases
```

#### 地区最新

```python
unique_provinces =  list(confirmed_df['Province/State'].unique())
# print(unique_provinces)
provinces_confirmed_cases = {}
for i in unique_provinces:
    cases = latest_confirmed[confirmed_df['Province/State']==i].sum()
    if cases > 0:
        provinces_confirmed_cases[i] = cases
    else:
        unique_provinces.remove(i)
```

```python
provinces_confirmed_cases = sorted(provinces_confirmed_cases.items(), key=lambda x:x[1], reverse=True)
provinces_confirmed_cases
```

```python
unique_countries =  list(confirmed_df['Country/Region'].unique())

country_confirmed_cases = []
no_cases = []
for i in unique_countries:
    cases = latest_confirmed[confirmed_df['Country/Region']==i].sum()
    if cases > 0:
        country_confirmed_cases.append(cases)
    else:
        no_cases.append(i)
        
for i in no_cases:
    unique_countries.remove(i)
    
# number of cases per country/region
for i in range(len(unique_countries)):
    print(f'{unique_countries[i]}: {country_confirmed_cases[i]} cases')
```

```python
unique_provinces =  list(confirmed_df['Province/State'].unique())

province_confirmed_cases = []
no_cases = [] 
for i in unique_provinces:
    cases = latest_confirmed[confirmed_df['Province/State']==i].sum()
    if cases > 0:
        province_confirmed_cases.append(cases)
    else:
        no_cases.append(i)
 
# remove areas with no confirmed cases
for i in no_cases:
    unique_provinces.remove(i)
    
# number of cases per province/state/city

for i in range(len(unique_provinces)):
    print(f'{unique_provinces[i]}: {province_confirmed_cases[i]} cases')
```

```python
nan_indices = [] 

# handle nan if there is any, it is usually a float: float('nan')

for i in range(len(unique_provinces)):
    if type(unique_provinces[i]) == float:
        nan_indices.append(i)

unique_provinces = list(unique_provinces)
province_confirmed_cases = list(province_confirmed_cases)

for i in nan_indices:
    unique_provinces.pop(i)
    province_confirmed_cases.pop(i)
```
