```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

# 散点图
matplotlib.pyplot.scatter()

```python
array1 = np.random.uniform(0.0, 5.0, size=(100, 2)) # 从均匀分布中随机采样
array1
```

```python
array2 = np.random.uniform(4.2, 10, size = (100, 2)) # 从均匀分布中随机采样
```

```python
# array1保存为txt，两列数据%f %f格式，每一列后'\n'换行
np.savetxt('./matplot_test_data1.txt', array1, fmt='%f %f', delimiter = '\n')
np.savetxt('./matplot_test_data2.txt', array2, fmt='%f %f', delimiter = '\n')
```

```python
# 读取txt
a = np.loadtxt('./matplot_test_data1.txt')
b = np.loadtxt('./matplot_test_data2.txt')
```

```python
# 绘制散点图
# plt.scatter（X列的数据， Y列的数据）
plt.scatter(a[ : , 0], a[ : , 1])
plt.scatter(b[ : , 0], b[ : , 1])
plt.show()
```

```python
# 定义x,y
x = [1, 2, 3 ,4]
y = [2, 4, 6, 8]
# scatter(s为点大小，color对应点的颜色，marker为点形状)
plt.scatter(x, y, s = [20, 40, 80, 320], color = ['r', 'g', 'b', 'y'], marker = 'h') # color='#ff0000'
```

```python
# 生成数据
male_high = np.random.uniform(140, 200, size = 20)
male_id = np.arange(21)  # 按序生成0-20
male_id = np.delete(male_id, 0)  # 删除0
male_high = male_high.reshape(20, 1)  # 一行转一列
male_id = male_id.reshape(20, -1)  # 一行转一列
male = np.concatenate((male_id, male_high), axis=1) # 按列连接high和id
```

```python
female_high = np.random.uniform(110, 185, size = 20)
female_id = np.arange(21)  # 按序生成0-20
female_id = np.delete(female_id, 0)  # 删除0
female_high = female_high.reshape(20, 1)  # 一行转一列
female_id = female_id.reshape(20, -1)  # 一行转一列
female = np.concatenate((female_id, female_high), axis=1) # 按列连接high和id
```

```python
X = np.concatenate((male, female), axis = 0)  # 按行连接male和female
y = np.concatenate((np.ones(20), np.zeros(20), axis = 0)  # 1代表男性 0代表女性
X[y == 1]  # 所有男性的行
X[y == 0]  # 所有女性的行
X[y ==1, 0] # 所有男性的行下取第0列，id
X[y ==1, 1] # 所有男性的行下取第1列，high
```

```python
data = np.concatenate((y.reshape(40, 1), X), axis = 1)
plt.scatter(data[y == 1, 1], data[y == 1, 2], color = 'r')
plt.scatter(data[y == 0, 1], data[y == 0, 2], color = 'g')
plt.show()
```

# 折线图
matplotlib.pyplot.plot()

```python
x = np.linspace(1, 20, 40)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label = 'sin(x)', color = 'r', linestyle = '--', marker='H')
plt.plot(x, y2, label = 'cos(x)', color = 'g', linestyle = '-', marker='h')
plt.show()
```

```python
x = np.linspace(1, 20, 40)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label = 'sin(x)', color = 'r', linestyle = '--', marker='H')
plt.plot(x, y2, label = 'cos(x)', color = 'g', linestyle = '-', marker='h')
plt.legend() # 加入提示
plt.show()
```

```python
x = np.linspace(1, 20, 40)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label = 'sin(x)', color = 'r', linestyle = '--', marker='H')
plt.plot(x, y2, label = 'cos(x)', color = 'g', linestyle = '-', marker='h')
plt.legend() # 加入提示
plt.axis([-5, 30, -2, 2]) # 图片缩小 axis[ X起点，X终点，Y起点，Y终点 ]
plt.show()
```

# 柱状图
matplotlib.pyplot.bar()

```python
x = ['Q1', 'Q2', 'Q3', 'Q4']
y = [10, 30, 20, 60]

rect = plt.bar(x, y, color = 'g', width = 0.3)
plt.grid(True)  # 加网格
# 绘制标度值
for index, item in enumerate(rect):  # enumerate()会产生一个index
    _x = item.get_x()
    _y = item.get_height()
    plt.text(_x, _y, y[index])
plt.ylim(0, 70)  # 限制y轴范围为0 - 70
plt.show()
```

# 饼图
matplotlib.pyplot.pie()

```python
x = ['Q1', 'Q2', 'Q3', 'Q4']
y = [10, 30, 20, 60]
plt.axes(aspect = 1)  # 正圆
plt.pie(y, labels = x, autopct = '%2.f%%', 
        explode = [0.2, 0, 0, 0], shadow=True)  # explode 抽出效果
plt.show()
```

# 直方图
matplotlib.pyplot.hist()

```python
np.random.seed(100)
data = np.random.normal(9000, 3000, size = 300)  # 正态分布生成，9000为均值，3000为标准差
# plt.hist( rwidth宽度，alpha透明度，bins柱子数 )
plt.hist(data, color = 'g', rwidth = 0.7, alpha = 0.6, bins = 20)  
plt.show()
```

# 箱线图
matplotlib.pyplot.boxplot()

```python
np.random.seed(100)
data = np.random.normal(size = 100)
plt.hist(data)
plt.show()
```

```python
data = np.concatenate([data, [4, 1 ,7 ,8, 9, -4]])
plt.boxplot(data)
plt.show()
```

```python
# whis：指定上下边界 四分位距离，默认1.5倍四分位差
plt.boxplot(data, whis = 3, showmeans = True, meanline = True)
plt.show()
```
